```python
n=int(input("Please enter n: "))
p=int(input("Please enter p: "))
m=int(input("Please enter m: "))
arr1=np.random.random((n,p))
arr2=np.random.random((p,m))
print(arr1)
print(arr2)
arr=np.zeros((n,m))
for i in range (n):
  for j in range(m):
    for k in range(p):
      arr[i,j]=arr1[i,k]*arr2[k,j]
print(arr)
```
